```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sigfig import round
pd.set_option('display.max_colwidth', None)
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
def p(val):
    return f"$ {round(str(val), decimals=0, spacing=3, spacer=',')}"

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

## Imputation

```python
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(train_X))
imputed_X_valid = pd.DataFrame(my_imputer.transform(val_X))

imputed_X_train.columns = train_X.columns
imputed_X_valid.columns = val_X.columns
```

## Tracked Imputation

```python
train_X_copy = train_X.copy()
val_X_copy = val_X.copy()

for col in cols_with_missing:
    train_X_copy[col + '_was_missing'] = train_X_copy[col].isnull()
    val_X_copy[col + '_was_missing'] = val_X_copy[col].isnull()

my_imputer = SimpleImputer()
imputed_train_X = pd.DataFrame(my_imputer.fit_transform(train_X_copy))
imputed_val_X = pd.DataFrame(my_imputer.transform(val_X_copy))

imputed_X_train.columns = train_X.columns
imputed_X_valid.columns = val_X.columns
```

```python
melbourne_model = DecisionTreeRegressor(random_state=1)
# Fit model
melbourne_model.fit(train_X, train_y)
```

```python
val_predictions = melbourne_model.predict(val_X)
p(mean_absolute_error(val_y, val_predictions))
```

```python
for max_leaf_nodes in [5, 50, 425, 450, 455, 460, 475, 500, 550, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes} \t\t Mean Absolute Error:", p(my_mae))
```

```python
forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
p(mean_absolute_error(val_y, melb_preds))
```

```python
output = pd.DataFrame({'Id': train_X.index,
                       'SalePrice': melb_preds})
output.to_csv('submission.csv', index=False)
```
